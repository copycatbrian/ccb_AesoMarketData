```python
import datetime as dt

def dayIter(startDate = dt.datetime(2017,1,1), endDate = dt.datetime.now()):
    curDate = startDate
    while curDate < endDate:
        yield curDate
        curDate = curDate +  dt.timedelta(days=1)
```
