```python
from polygon.rest import RESTClient
```

```python
import datetime as dt
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import plot
```

```python
client = RESTClient(config.api_key) # api_key is used

stockTicker = 'AAPL'

# daily bars
dataRequest = client.get_aggs(ticker = stockTicker, 
                              multiplier = 1,
                              timespan = 'day',
                              from_ = '2000-01-01',
                              to = '2100-01-01')
```

```python
dataRequest
```

```python
priceData = pd.DataFrame(dataRequest)

#create Date column
priceData['Date'] = priceData['timestamp'].apply(
                          lambda x: pd.to_datetime(x*1000000))

priceData = priceData.set_index('Date')

#generate plotly figure
fig = go.Figure(data=[go.Candlestick(x=priceData.index,
                open=priceData['open'],
                high=priceData['high'],
                low=priceData['low'],
                close=priceData['close'])])

#open figure in browser
plot(fig, auto_open=True)
```

```python
exchanges = pd.DataFrame(client.get_exchanges(asset_class='stocks',
                                              locale='us'))

# # exchanges MIC for use in list tickers
# print(exchanges.mic)

# # remove duplicates and remove None
exchangeList = list(set(exchanges.mic))
exchangeList.remove(None)

# get all tickers from all US exchanges
usTickers = []
for x in exchangeList:
    # page through response
    for t in client.list_tickers(market='stocks',
                                 exchange=x,
                                 active=False,
                                 limit=20):
        # add to list
        usTickers.append(t.ticker)
    # print exchange when finished   
    print(x)
        
# final ticker list
finalTickerList = set(usTickers)
```
