```python
    #### datafolder is where all the data is stored 


    #### datafolder is where all the data is stored 




#### Reads and extracts ActualForecast reports from folder 
i = 0 
filenames = [f for f in os.listdir(subfolders2[i]) if '.csv' in f]
filelist = os.listdir(subfolders2[i])

selectedReports = [f for f in filelist if '.csv' in f]
selectedReports

rptDir = subfolders2[i]
reportName = rptDir.split(sep="\\")[-1]
reportName


dataDF = pd.DataFrame()

for rpts in selectedReports:
    ### make path  
    fpath = rptDir + "/" +rpts
    #print(fpath)
    tmp = pd.read_csv(fpath, sep=',',  skiprows=2)

    dataDF = pd.concat([dataDF,tmp])



dataDF.to_csv(subfolders[0] + "/"+ reportName +"-ending2022.csv" )     
```
