```python
colTypes  = ['Asset Id', 'MC', 'Type'] 
csdTypes = csdTypes.loc[:,colTypes]
colTypes  = ['AssetID', 'MC', 'Type'] 
aesoTypes = aesoTypes.loc[:,colTypes]

csdTypes.columns = colTypes
aesoTypes.columns = colTypes
```

```python
FuelTypes =  pd.merge(aesoTypes, csdTypes, on= 'AssetID', how='outer')
```

```python
mask = FuelTypes['Type_x'].isna()
FuelTypes.loc[mask, ['MC_x',"Type_x"]] = FuelTypes.loc[mask, ['MC_y',"Type_y"]].values
```

```python
FuelTypes
```

```python
df['PriceMW'] = df['Price'] * df['Available MW']
df['CapturePMW'] = df['Price'] * df['Dispatched MW']
```

```python
curUnit =  moUnits[0]
for curUnit in moUnits:
    mask = df['Asset Id'] == curUnit        
    try:
        df.loc[mask,'FuelType'] = FuelTypes.loc[FuelTypes['AssetID'] == curUnit].iloc[0,2]
        print(FuelTypes.loc[FuelTypes['AssetID'] == curUnit].iloc[0,2])
    except:
        df.loc[mask,'FuelType'] = 'Unlisted'
        print("Unlisted")
```
